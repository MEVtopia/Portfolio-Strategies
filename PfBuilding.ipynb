```python
%%time

price_df_list = []
for i in tqdm(range(120)):
    price_data = data.iloc[:,3*i:3*i+2].copy()
    price_data.dropna(inplace=True)
    price_data.iloc[:,0] = pd.to_datetime(price_data.iloc[:,0], unit= 'D', origin= '1899-12-30')
    price_data.rename(columns = {price_data.columns[0] : 'Date', price_data.columns[1]: price_data.columns[0]}, inplace = True)
    price_data.set_index('Date', inplace = True)
    price_df_list.append(price_data)

agg_price_data = price_df_list[0]
for df in tqdm(price_df_list[1:]):
    agg_price_data = agg_price_data.join(df, how='outer', sort = True)

agg_price_data = agg_price_data[(agg_price_data.index >= '31/12/2018') & (agg_price_data.index <= '31/12/2020')]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FuncFormatter



class Strategy:

    def __init__(self, name, returns):
        self.name = name
        self.returns = returns
        self.pnl = np.c_[np.array([1000000.]), 1000000.*np.cumprod(1.+returns.to_numpy()).reshape(1,-1)].flatten()


    def volatility(self):
        return np.std(self.returns)*np.sqrt(252)

    def sharpe_ratio(self):
        return np.mean(self.returns)*np.sqrt(252)/np.std(self.returns)

    def max_dd(self):
        return 1. - np.min(np.flip(np.minimum.accumulate(np.flip(self.pnl)))/self.pnl)

    def max_dd_2(self):
        return np.max(1. - self.pnl/np.maximum.accumulate(self.pnl))

    def illustrate(self):
        fig, ax = plt.subplots(1, 1)
        ax.plot(self.returns.index, self.pnl[1:], 'b', label=self.name + ' (vol: %.2f, Sharpe: %.2f, MDD: %.2f)'%
                                                      (self.volatility(), self.sharpe_ratio(), self.max_dd()))
        ax.set_xlabel('Time')
        ax.set_ylabel('PnL')
        ax.legend()
        ax.tick_params(axis='x', rotation=45)
        ax.set_title(f'PnL of the strategy {self.name}')
        ax.grid(axis='x', linestyle='--')

        # Define a custom y-axis tick label formatter
        def y_formatter(x, pos):
            if x >= 1_000_000:
                return f"{x*1e-6:.0f}M"
            elif x >= 1_000:
                return f"{x*1e-3:.0f}K"
            else:
                return f"{x:.0f}"

        ax.yaxis.set_major_formatter(FuncFormatter(y_formatter))

        fig.tight_layout()
        fig.savefig(f'{self.name}.pdf')
        plt.show()


class CapiWeighted(Strategy):

    def __init__(self, name, returns, capitalizations):
        weights_CW = capitalizations.copy()
        weights_CW['sum'] = weights_CW.sum(axis=1)
        weights_CW = weights_CW.div(weights_CW['sum'], axis=0)
        weights_CW.drop(columns='sum', inplace=True)
        weights_CW = weights_CW.shift(1).dropna()
        returns_CW = (returns * weights_CW).sum(axis=1)
        super().__init__(name, returns_CW)

class EqualWeighted(Strategy):
    def __init__(self, name, returns):
        # Calculate equal weights for each asset in the portfolio
        weights_EW = pd.DataFrame(1.0/len(returns.columns), index=returns.index, columns=returns.columns)
        
        # Compute portfolio returns using equal weights
        returns_EW = (returns * weights_EW).sum(axis=1)
        
        # Initialize parent Strategy class with the computed returns
        super().__init__(name, returns_EW)

class Markowitz(Strategy):
    def __init__(self, name, returns, cov_matrix):
        # Compute the inverse of the covariance matrix
        inv_cov_matrix = np.linalg.inv(cov_matrix)
        
        # Create a ones vector with the same length as the number of assets
        e = np.ones(len(returns.columns))
        
        # Compute the Markowitz optimal weights
        w = inv_cov_matrix.dot(e) / (e.T.dot(inv_cov_matrix).dot(e))
        
        # Compute portfolio returns using the Markowitz weights
        returns_MW = (returns * w).sum(axis=1)
        
        # Initialize parent Strategy class with the computed returns
        super().__init__(name, returns_MW)

class ERC(Strategy):
    def __init__(self, name, returns, cov_matrix): 
        # Extract variances from the covariance matrix
        variance = np.diag(cov_matrix)
        
        # Compute inverse of the variances
        inv_var = 1 / variance
        
        # Compute the Equal Risk Contribution (ERC) weights
        w = inv_var / np.sum(inv_var)
        
        # Compute portfolio returns using the ERC weights
        returns_ERC = (returns * w).sum(axis=1)
        
        # Initialize parent Strategy class with the computed returns
        super().__init__(name, returns_ERC)

class ComparisonStrategy:
    def __init__(self, strategies):
        # Store the list of strategies to be compared
        self.strategies = strategies
        
    def illustrate(self):
        # Create a new plot
        fig, ax = plt.subplots(1, 1)
        
        # Plot each strategy's PnL on the same graph
        for strategy in self.strategies:
            ax.plot(strategy.returns.index, 
                    strategy.pnl[1:], 
                    label=f'{strategy.name} (vol: {strategy.volatility():.2f}, Sharpe: {strategy.sharpe_ratio():.2f}, MDD: {strategy.max_dd():.2f}')
            
        # Set plot labels, title, and display settings
        ax.set_xlabel('Time')
        ax.set_ylabel('PnL')
        ax.legend()
        ax.tick_params(axis='x', rotation=45)
        ax.set_title('PnL Comparison of Strategies')
        ax.grid(axis='x', linestyle='--')
        fig.tight_layout()
        plt.show()
```
